```python
file_path = '/content/dataLaterality.csv'


df = pd.read_csv(file_path)

print("First five rows of all columns:")
print (df.head())
```

```python
print("First five rows of all columns:")
print(df.head())
```

```python

correlation_matrix = df[['cancer', 'implant','age', 'Laterality_encoded']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Cancer vs Implant vs Age vs Laterality')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(x='implant', y= 'Laterality_encoded' , hue='cancer', data=df, palette='viridis')
plt.title('Cancer Probability by Implant and Laterality Encoded')
plt.xlabel('Implant')
plt.ylabel('Laterality Encoded')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(x='Laterality_encoded', y= 'age' , hue='cancer', data=df, palette='viridis')
plt.title('Cancer Probability by Age and Laterality Encoded')
plt.xlabel('Laterality_encoded')
plt.ylabel('Age')
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))

data =[df['age'] ,df['implant'],df['cancer'], df['Laterality_encoded']]

ax.boxplot(data)

ax.set_xticklabels(['age', 'implant','cancer','Laterality_encoded'])
ax.set_yticklabels('values')
ax.set_title('Box Plot for Four Parameters')

plt.show()
```
